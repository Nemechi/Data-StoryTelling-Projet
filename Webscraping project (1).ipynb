```python
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup as bs
import numpy as np
```

```python
def scrape_reviews():
    review_date = []
    review_text = []
    ratings = []
    
    base_url = "https://www.consumeraffairs.com/automotive/toyota-camry.html"

    for i in range(1,14): # fetching reviews from thirteen pages
        current_url = f"{base_url}?page={i}"
        data = requests.get(current_url)
        soup = bs(data.text, 'html.parser')
        all_reviews = soup.find_all('div', {'class': 'rvw js-rvw'}) #all reviews on ith page
        for review in all_reviews:
            rvw_data = review.find('div', {'class': 'rvw-bd'}).find('span').getText().replace('Original review: ', '')
            review_date.append(rvw_data)
            # review_body = review.find('p').getText()
            review_body = review.find('div', {'class': 'rvw-bd'}).find('p').getText()
            review_text.append(review_body)
        
            img = review.find('div', {'class': 'rvw__hdr'}).find('img')
            if img:
                rating = img.get('data-rating')
            else:
                rating = np.nan
            ratings.append(rating) 
        
        time.sleep(1) #delay for 1 second before going to next page
    
    data_dict = {'date': review_date, 'review_text': review_text, 'star_rating': ratings}
    return pd.DataFrame(data_dict)

#calling the function
ReviewDf = scrape_reviews()
ReviewDf
```

```python
ReviewDf.to_csv('toyota_data.csv', index = False)
```
